## Описание проекта

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Требуется построить модель, которая предскажет температуру стали.

**Описание этапа обработки**

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# загрузка данных
def data_load(path_serv, path_comp):
    try:
        df = pd.read_csv(path_serv)
    except:
        df = pd.read_csv(path_comp)
    display(df.head())
    display(df.info())
    display(df.describe(include='all'))
    print(f'Количество дубликатов: {df.duplicated().sum()}')
    return df
```

```python
# пропущенные значения бары
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by='space', ascending=True)
            .plot(kind = 'barh', figsize=(12,4), legend=False, grid=True)
            .set_title(f'Процент пропущенных значений', fontsize=16)    
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции')
```

```python
# проверка распределений
def charts(df, col=None):
    if col is None:
        col = df.columns
    for i in col:
        f, axes = plt.subplots(1, 2, figsize=(17,5))
        axes[0].set_title(f'Гистограмма по {i}', fontsize=15)
        axes[0].set(xlabel=i, ylabel='Количество')
        axes[1].set_title(f'Диаграмма размаха по {i}', fontsize=15)
        sns.histplot(df[i], kde=True, bins=50, ax=axes[0])
        axes[0].grid() 
        sns.boxplot(x=df[i], ax=axes[1])
        plt.grid()
        plt.show()
```

## Предобработка данных

### Данные об электродах

```python
df_arc_new.head()
```

```python
bad_keys = df_arc_new.loc[df_arc_new['Реактивная мощность']<0, 'key'] 
df_arc_new = df_arc_new.loc[~df_arc_new['key'].isin(bad_keys)]
```

```python
df_arc_new['Начало нагрева дугой'] = pd.to_datetime(df_arc_new['Начало нагрева дугой'], format='%Y-%m-%d %H:%M:%S')
df_arc_new['Конец нагрева дугой'] = pd.to_datetime(df_arc_new['Конец нагрева дугой'], format='%Y-%m-%d %H:%M:%S')
```

### Создание выборок

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor
from sklearn.pipeline import Pipeline
from sklearn import set_config
from catboost import CatBoostRegressor
from tqdm import notebook
```

```python
RANDOM_STATE = 40823
CV = 3
scorer = 'neg_mean_absolute_error'
```

```python
features = df.drop('last_temp', axis=1)
targets = df['last_temp']

features_train, features_test, targets_train, targets_test = train_test_split(features, 
                                                                              targets, 
                                                                              test_size=0.25, 
                                                                              random_state=RANDOM_STATE)

features_train.shape, features_test.shape, targets_train.shape, targets_test.shape
```

### Обучение моделей

```python
dummy_regr = DummyRegressor()
dummy_regr.fit(features_train, targets_train)
pred = dummy_regr.predict(features_test)
print(f'DummyRegressor MAE = {mean_absolute_error(targets_test, pred):.3f}')
```

```python
# пайплайны для LinearRegression, RandomForestRegressor, CatBoostRegressor
pipe_lr = Pipeline([('stsc', StandardScaler()), 
                    ('lr', LinearRegression())])

pipe_rfr = Pipeline([('stsc', StandardScaler()),
                      ('rfr', RandomForestRegressor(random_state=RANDOM_STATE))])

pipe_cbr = Pipeline([('stsc', StandardScaler()),
                     ('cbr', CatBoostRegressor(loss_function='MAE', random_state=RANDOM_STATE, verbose=False))])
# подбираемые в GridSearchCV и RandomizedSearchCV гиперпараметры
grid_params_lr = [{'lr__fit_intercept': [True, False]}]

grid_params_rfr = [{'rfr__n_estimators': [50, 100],
                    'rfr__max_depth': range(2, 8), 
                    'rfr__min_samples_split': range(2, 8), 
                    'rfr__min_samples_leaf': range(1, 8)}]

grid_params_cbr = [{'cbr__n_estimators': [50, 100], 
                    'cbr__max_depth': range(2, 8)}]
# создание GridSearchCV и RandomizedSearchCV для разных моделей
LR = GridSearchCV(estimator=pipe_lr, 
                  cv=CV, 
                  param_grid=grid_params_lr, 
                  scoring=scorer)

RFR = RandomizedSearchCV(estimator=pipe_rfr, 
                         cv=CV,
                         param_distributions=grid_params_rfr, 
                         scoring=scorer, 
                         n_iter=50, 
                         random_state=RANDOM_STATE)

CBR = RandomizedSearchCV(estimator=pipe_cbr, 
                         cv=CV, 
                         param_distributions=grid_params_cbr, 
                         scoring=scorer, 
                         n_iter=6, 
                         random_state=RANDOM_STATE)
# список пайплайнов для итерации
grids = [LR, RFR, CBR]
grids_dict = {0: 'LinearRegression', 
              1: 'RandomForestRegressor', 
              2: 'CatBoostRegressor'}
# худшее значение MAE
best_mae = 6.8
```

```python
%%time
for idx, gs in enumerate(grids):
    print('_' * 20)
    print(f'Модель: {grids_dict[idx]}')
    # обучаем модель
    gs.fit(features_train, targets_train)
    # параметры лучшей модели
    print(f'Лучшие параметры: {gs.best_params_}')
    print(f'Лучшее значение MAE: {abs(gs.best_score_):.3f}')
    if abs(gs.best_score_) < best_mae:
        best_mae = abs(gs.best_score_)
        best_gs = gs
        best_gs_idx = idx
print('_' * 20)      
print(f'Лучшая модель: {grids_dict[best_gs_idx]}')
pred = best_gs.predict(features_test)
print(f'Значение MAE на тестовой выборке: {mean_absolute_error(targets_test, pred):.3f}')
```
