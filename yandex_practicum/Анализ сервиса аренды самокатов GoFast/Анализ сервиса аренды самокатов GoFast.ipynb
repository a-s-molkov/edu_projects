## Описание проекта

Получены данные о некоторых пользователях из нескольких городов, а также об их поездках. Требуется проанализировать данные и проверить некоторые гипотезы, которые могут помочь бизнесу вырасти.
Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться:
* без подписки
    * абонентская плата отсутствует;
    * стоимость одной минуты поездки — 8 рублей;
    * стоимость старта (начала поездки) — 50 рублей;
* с подпиской Ultra
    * абонентская плата — 199 рублей в месяц;
    * стоимость одной минуты поездки — 6 рублей;
    * стоимость старта — бесплатно.

## Описание данных

В основных данных есть информация о пользователях, их поездках и подписках.

***Пользователи — users_go.csv***:
* **user_id** - уникальный идентификатор пользователя
* **name** - имя пользователя
* **age** - возраст
* **city** - город
* **subscription_type** - тип подписки (free, ultra)

***Поездки — rides_go.csv***:
* **user_id** - уникальный идентификатор пользователя
* **distance** - расстояние, которое пользователь проехал в текущей сессии (в метрах)
* **duration** - продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»
* **date** - дата совершения поездки

***Подписки — subscriptions_go.csv***:
* **subscription_type** - тип подписки
* **minute_price** - стоимость одной минуты поездки по данной подписке
* **start_ride_price** - стоимость начала поездки
* **subscription_fee** - стоимость ежемесячного платежа

## Подсчёт выручки

```python
df_common.head()
```

```python
# создание датафрейма с агрегированными данными о поездках
df_rides_agg = df_common.pivot_table(
    index=['subscription_type','minute_price','start_ride_price','subscription_fee','user_id','month'], 
    values=['distance','duration'], 
    aggfunc={'distance':'sum', 'duration':['sum','count']})
df_rides_agg.columns = ['distance_sum','count','duration_sum']
df_rides_agg = df_rides_agg.reset_index()
df_rides_agg.sample(20)
```

```python
# добавление столбца с ежемесячной выручкой
df_rides_agg['monthly_revenue'] = (df_rides_agg['start_ride_price'] * df_rides_agg['count']
                                   + df_rides_agg['minute_price'] * df_rides_agg['duration_sum'].round()
                                   + df_rides_agg['subscription_fee'])
df_rides_agg.sample(20)
```

```python
# Код ревьюера
df_rides_agg.groupby('subscription_type')['monthly_revenue'].mean()
```

## Проверка гипотез

### Тратят ли пользователи с подпиской больше времени на поездки?

<div style="background-color:black; color:#ffffff; text-align:left; padding:2%">
    
Нулевая гипотеза: пользователи с подпиской и без подписки тратят одинаковое время.
    
Альтернативная гипотеза: пользователи с подпиской тратят больше времени на поездки.

```python
result = st.ttest_ind(df_ultra['duration'],
                      df_free['duration'], 
                      alternative='greater',
                      equal_var=False)
print('p-value:', result.pvalue)
alpha = 0.05
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

<div style="background-color:black; color:#ffffff; text-align:left; padding:2%">
    
Пользователи с подпиской тратят больше времени на поездки.

### Расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? 

<div style="background-color:black; color:#ffffff; text-align:left; padding:2%">
    
Нулевая гипотеза: расстояние, которое проезжают пользователи с подпиской за одну поездку равно 3130 метров.
    
Альтернативная гипотеза: расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров.

```python
result = st.ttest_1samp(df_ultra['distance'],
                      3130, 
                      alternative='less')
print('p-value:', result.pvalue)
alpha = 0.05
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

### Будет ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки?

<div style="background-color:black; color:#ffffff; text-align:left; padding:2%">
    
Нулевая гипотеза: выручки от пользователей с подпиской и без неё равны.
    
Альтернативная гипотеза: помесячная выручка от пользователей с подпиской по месяцам выше.

```python
result = st.ttest_ind(df_rides_agg.loc[df_rides_agg['subscription_type']=='ultra','monthly_revenue'],
                      df_rides_agg.loc[df_rides_agg['subscription_type']=='free','monthly_revenue'],
                      alternative='greater',
                      equal_var=False)
print('p-value:', result.pvalue)
alpha = 0.05
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

<div style="background-color:black; color:#ffffff; text-align:left; padding:2%">
    
Помесячная выручка от пользователей с подпиской по месяцам выше.
